```python
# 씨본로드
import seaborn as sns
```

```python
# 뒷배경
sns.set_style('whitegrid')
```

```python
# 기본 데이터
# tips : 요일별 점심, 저녁, 흡연여부, 식사금액, 팁 금액, 성별, 인원수
tips = sns.load_dataset( 'tips' ) # ex) 금요일날 팁을 잘주는지, 흡연자가 팁을 잘주는지 등등을 알 수 있게금 한다
tips # 결과값을 보니 pivot 작성의 냄새가 난다
```

```python
# 데이터 표현( 결과는 횡으로 표현됨 )
# 가격대가 어디에 몰려 있는지
plt.figure( figsize=( 8, 6 ) )
sns.boxplot( x=tips['total_bill'] ) # x는 tips의 total_bill을 표현, 실제그리기는 seaborn으로 그린다
plt.show() 

# 박스플롯의 왼쪽이 최소, 오른쪽이 최대, 파란색박스안의 선은 중간값( 대략 18로 보임 )
```

```python
tips.columns
```

```python
plt.figure( figsize=( 8, 6 ) )
# x축과 y축의 컬럼을 지정, 데이터는 별도로 대입한다
sns.boxplot( x='day', y='total_bill', data=tips ) 
plt.show() 
```

```python
plt.figure( figsize=( 8, 6 ) )
# 위의 시각화 자료에 흡연여부를 넣겠다
sns.boxplot( x='day', y='total_bill', hue='smoker', data=tips, palette='Set3' ) # hue로 인해 시각화자료가 나눠지기 시작함, platte: 컬러지정
plt.show() 
```

```python
plt.figure( figsize=( 8, 6 ) )
# 무리형태로 -> 포인트 중심 시각화
sns.boxplot( x='day', y='total_bill', data=tips, palette='Set3' )
sns.swarmplot( x='day', y='total_bill', data=tips, color='0.5' ) # swarmplot : 포인트로 표현
plt.show() 
```

### 산포도 + 평균선(x,y에 대한 1차 함수)

```python
sns.set_style('darkgrid')
sns.lmplot( x='total_bill', y='tip', data=tips, size=7 ) # 매출대비 팁의 현황( 음식을 어느정도 먹으면 팁을 어떻게 주는가 )
plt.show()
```

```python
sns.lmplot( x='total_bill', y='tip', hue='smoker', data=tips, size=7 ) 
plt.show()
```

```python
# matplotlib에 비해 코드량이나 작업량이 훨씬 간결하고 파워풀하다
sns.lmplot( x='total_bill', y='tip', hue='smoker', data=tips, size=7, palette='Set1' ) 
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights.shape
```

```python
flights.head(2)
```

```python
flights['year'].unique(), len(flights['year'].unique())
```

```python
flights['month'].unique(), len(flights['month'].unique())
```

```python
# index에 12개월을 집어넣고, columns에 년도, data는 고객수를 집어넣는다
flights_piv = flights.pivot( 'month', 'year', 'passengers' ) # ( 'index', 'columns', 'data' )
flights_piv
```

```python
# 히트맵을 통해서 고객수가 어느 방향(년도, 월)으로 증가하고 있는지, 기세를 확인해본다
plt.figure( figsize=( 10, 8 ) )
sns.heatmap( flights_piv )
plt.show()
```

```python
plt.figure( figsize=( 10, 8 ) )
# 데이터 세팅
# annot : 숫자 세팅
# fmt : format줄임말, 수치의 타입, d: 정수
sns.heatmap( flights_piv, annot=True, fmt='d' )
plt.show()

# 이런 결과물은 지역구별의 대선투표율 정도로 표현도 가능하다
```

```python
# 산포 행렬
iris = sns.load_dataset( 'iris' )
iris.head(2)
```

```python
# 대각선 기준으로 대칭으로 확인
sns.pairplot( iris, hue='species' )
plt.show() # 결과는 iris의 변수가 4개이기 때문에 4 x 4밖에 안나온다
```

```python
sns.pairplot( iris, hue='species', vars=['sepal_width', 'sepal_length' ] ) # vars : variables
plt.show()
```

```python
iris.columns
```

```python
sns.pairplot( iris, hue='species', x_vars=['sepal_width', 'sepal_length' ],
            y_vars=['petal_width', 'petal_length']) # vars : variables
plt.show()
```
